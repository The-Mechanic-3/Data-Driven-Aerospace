```python
import numpy as np
import matplotlib.pyplot as plt

# Generate a random 50x50 matrix
matrix = np.random.normal(size=(50, 50))

# Compute the SVD
U, S, V = np.linalg.svd(matrix)

# Plot the singular values
plt.plot(S)
plt.title("Singular Values")
plt.xlabel("Index")
plt.ylabel("Singular Value")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Number of repetitions
num_repetitions = 100

# Store singular values from all repetitions
all_singular_values = []

# Repeat SVD computation
for _ in range(num_repetitions):
    # Generate a random matrix
    matrix = np.random.normal(size=(50, 50))

    # Compute the SVD
    U, S, V = np.linalg.svd(matrix)

    # Store singular values
    all_singular_values.extend(S)

# Create box-and-whisker plot
plt.boxplot(all_singular_values)
plt.title("Distribution of Singular Values")
plt.ylabel("Singular Value")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Number of repetitions
num_repetitions = 100

# Store mean and median singular values
mean_singular_values = []
median_singular_values = []

# Repeat SVD computation
for _ in range(num_repetitions):
    # Generate a random matrix
    matrix = np.random.normal(size=(50, 50))

    # Compute the SVD
    U, S, V = np.linalg.svd(matrix)

    # Store mean and median singular values
    mean_singular_values.append(np.mean(S))
    median_singular_values.append(np.median(S))

# Plot mean and median singular values
plt.plot(mean_singular_values, label="Mean")
plt.plot(median_singular_values, label="Median")
plt.title("Mean and Median Singular Values")
plt.xlabel("Repetition Index (r)")
plt.ylabel("Singular Value")
plt.legend()
plt.show()
```
